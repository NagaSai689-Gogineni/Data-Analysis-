```python
# 2) Can you tell me the Distribution of Vehicles based on Year of Registration with the help of a plot
# Create a histogram
plt.figure(figsize=(12, 6))
plt.hist(data['yearOfRegistration'], bins=20, edgecolor='k')
plt.title("Distribution of Vehicles Based on Year of Registration")
plt.xlabel("Year of Registration")
plt.ylabel("Number of Vehicles")
plt.show()
```

```python
#creating countplot
plt.figure(figsize=(12, 6))
sns.countplot(x='yearOfRegistration', data=data, palette='viridis')
plt.title("Distribution of Vehicles Based on Year of Registration")
plt.xlabel("Year of Registration")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=90)
plt.show()
```

```python
a=pd.pivot_table(data1, index="vehicleType", values="price", aggfunc="mean")
a
```

```python
#Create a plot based on the Variation of the price range by the vehicle type
sns.barplot(a,x=a.index, y="price")
plt.title("Variation of the price range by the vehicle type\n")
plt.xticks(rotation=90)
plt.show()
```

```python
#4) Find out Total count of vehicles by type available on ebay for sale.As well as create a visualization for the client
data1["vehicleType"].value_counts()    
```

```python
sns.countplot(data1, x="vehicleType")
plt.title("Total count of vehicles by type available on ebay for sale\n")
plt.xticks(rotation=90)
plt.show()
```

```python
# 5) Is there any relationship between dollar_price and kilometer? 
data1[["price", "kilometer"]].corr(method="pearson")
```

```python
sns.heatmap(data1[["price", "kilometer"]].corr(method="pearson"), annot=True)
plt.title("Relationship between dollar_price and kilometer\n")
plt.show()
```
